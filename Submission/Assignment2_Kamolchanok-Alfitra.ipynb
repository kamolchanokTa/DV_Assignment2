##  Q1: What is the majority of  handedness among baseball players?

To represent this data we are showing a pie chart. The handedness values, left, right and both are identified and their count is calculated from the data. From this count we can generate a pie chart. 

```python
#Q1: What is the dominant handedness of baseball players?

# compute a histogram of the handedness variable
df_handedness = df['handedness'].value_counts().reset_index()
# set column names
df_handedness.columns = ['handedness', 'count']
df_handedness.head()
```

```python
#Q1: Contd.

# display pie chart of handedness
import matplotlib.pyplot as plt

#For labels
handedness_labels = ['Right','Left','Both']

plt.figure(figsize=(4, 4))

colors = ['#2171b5','#4292c6','#c6dbef']

#draw pie chart
plt.pie(df_handedness['count'], labels=handedness_labels, colors=colors,
        autopct='%1.1f%%', startangle=90)
plt.title('Handedness count of baseball players', y=1.1)
plt.axis('equal')
plt.show()
```

From the pie chart we can see that the majority of handedness is Right at 63.7%. The number of right handed players are more than twice the number of left handed players. Players who both hands are identified also significant at 9%.

##  Q2: What is the average of the batting averages by handedness?
We decided to use a bar chart to show the mean of batting averages grouped by handedness.

```python
#Q2: What is the average of the batting averages by handedness?
import numpy as np
from bokeh.io import show
from bokeh.plotting import figure

# x-axis: handedness
handedness = ['Both','Left','Right']

# y-axis: avg - mean of average batting
mean_avg = df.groupby('handedness').mean()['avg']

colors = ['#c6dbef','#4292c6','#2171b5']
q2 = figure(x_range=handedness, y_range=(0,0.23), plot_width = 400, plot_height=300, 
            x_axis_label= "Handedness", y_axis_label="Average batting",
            title="Average of Batting Average")
q2.vbar(x=handedness, top=mean_avg, width=0.4, color=colors)

# to make the x-grid invisible
q2.xgrid.grid_line_color = None

show(q2)
```

This bar chart shows the mean of batting average grouped by handedness. We can see that the mean of batting average among the players who are left-handed and both-handed is almost similar. However, the mean among the players who are right-handed is slightly lower than the other two groups although they are the majority of baseball players in this study (based on the finding in Q1.)

```python
#Q2: What is the average of home runs by handedness?
import numpy as np
from bokeh.core.properties import value
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import LinearAxis
from bokeh.transform import dodge


# x-axis: handedness
handedness = ['Both','Left','Right']

# y-axis: avg - mean of average batting
mean_avg = df.groupby('handedness').mean()['avg']
print(mean_avg)
# y-axis-Right: avg - mean of average batting
mean_hr = df.groupby('handedness').mean()['HR']
print(mean_hr)
data = {'handedness' : handedness,
        'HR'   : mean_hr,
        'avg'   : mean_avg}
source = ColumnDataSource(data=data)
colors = ['#c6dbef','#4292c6','#2171b5']
colors_avgbat = ['#e5f5f9','#99d8c9','#2ca25f']
q2 = figure(x_range=handedness, plot_width = 500, plot_height=300, 
            x_axis_label= "Handedness", y_axis_label="Average number of Home Runs",
            title="Average number of Home Runs",)
q2.vbar(x=dodge('handedness', -0.25, range=q2.x_range), top='HR', width=0.2, source=source,
       color="#c9d9d3", legend=value("HR"))

q2.vbar(x=dodge('handedness',  0.0,  range=q2.x_range), top='avg', width=0.2, source=source,
       color="#718dbf", legend=value("avg"),y_range_name='Average of Batting Average')
q2.extra_y_ranges = {"Average of Batting Average": Range1d(start=0, end=0.23)}
q2.add_layout(LinearAxis(y_range_name="Average of Batting Average",axis_label='Average of Batting Average'), 'right' )

# to make the x-grid invisible
q2.xgrid.grid_line_color = None

show(q2)
```

## Q4.How does the height of a baseball player affect his batting average?

For this query we will look at the heights of the players and their respective batting averages by drawing a scatter plot. We try to see how the height of players affect the performance indicated by their batting averages. Additionally we will identify the height of the player with the highest batting average. 

```python
#Q4: Does the batting average correlate to Height?
from bokeh.models import BoxAnnotation, Span

#x-axis: height of players
df_heights = df['height']

#y-axis: batting averages
df_batting_avg = df['avg']

q4 = figure(plot_width=650, plot_height=400, x_axis_label= "Player heights (inches)", y_axis_label="Batting average",
           title="Batting average distribution with player heights")

# circles for the distribution
q4.circle(df_heights, df_batting_avg, size=6, color="navy", alpha=0.5)
q4.xaxis[0].ticker.desired_num_ticks = 15

#highest batting average
highest_avg = df['avg'].max()
q4.circle(72, highest_avg, size=6, color="red")

#hightlight the best batting averages 
best_box = BoxAnnotation(plot=q4, bottom=0.3, left=69, right=75, fill_alpha=0.25, fill_color='yellow')

#line for height mean
vline = Span(location=72.756266, dimension='height', line_color='red', line_width=1)

q4.renderers.extend([best_box, vline])

# show the results
show(q4)
```

With this scatter plot of the batting averages with the player heights, we can see that the batting averages are high (more than 0.3) for the height range between 69 and 75 inches (highlighted in yellow area). The player with highest batting average has a height of 72 inches (shown as a red dot). If we consider the mean of heights, 72.756266, players with heights close to this mean usually have high batting averages.
In the next bar chart we look at how the batting average mean is distributed with the player heights.

```python
# Q4: Contd.

#average batting averages grouped by height
mean_avg_by_height = df.groupby('height').mean()['avg']
df_height_meanavg = mean_avg_by_height.reset_index()

q4 = figure(x_range=(64, 81), y_range=(0,0.3), plot_width = 600, plot_height=400, 
            x_axis_label= "Player heights", y_axis_label="Mean Batting average",
            title="Average of Batting Average over height")
q4.vbar(x=df_height_meanavg['height'], top=mean_avg_by_height, width=0.4)

#increase x axis ticks
q4.xaxis[0].ticker.desired_num_ticks = 15

show(q4)
```

This chart is left skewed indicating that the mean of batting average is higher for shorter players. Specifically, the player height 67 has a mean batting average of 0.263250. And there is a decline of batting averages with the increase of height.
However, This can be attributed to the fact of having a smaller number of short players. Nonetheless, even those amount of players have managed to get a high batting average among themselves. Even in the scatter plot we see that the players shorter than 69 have managed to keep a non-zero batting average.

## Q5.Does the batting average correlate to the number of home runs?
In order to identify the correlation between batting average and the number of home runs. We decided to use a scatter plot.

```python
#Q5: Does the batting average correlate to the number of home runs?
from bokeh.models import LinearAxis, Range1d

# x-axis: avg - batting average
df_avg = df['avg']

# y-axis: HR - number of home runs
df_hr = df['HR']

q5 = figure(plot_width=800, plot_height=600, x_axis_label= "Batting average", y_axis_label="# Home runs",
           title="Relationship between batting average and the number of home runs")

# add a circle renderer with a size, color, and alpha
q5.circle(df_avg, df_hr, size=5, color="navy", alpha=0.3)

#hightlight the best batting averages 
best_box = BoxAnnotation(plot=q5, bottom=200, left=0.2, right= 0.31, fill_alpha=0.25, fill_color='yellow')
q5.renderers.extend([best_box])

# show the results
show(q5)
```

From the scatter plot, we can see that most baseball players have batting average between 0.21 and 0.28. (The mean of batting average is 0.19.) However, we do not see any correlation between batting average and the number of home runs. It is noticeable that most players who have larger number of home runs (more than 200) are usually have batting average between 0.25 and 0.3 which is higher than the mean of batting average.
Although batting average is a performance indicator, it does not determine the number of home runs. In other words, it cannot be concluded that the players who have higher batting average will always have higher change of hitting a home run.

## Q6.Who are the top ten best baseball players judging from the average batting and the number of home runs?
We first looked at the top 10 players according to batting average. The data is sorted with descending order of batting average and the top 10 players are filtered. Their number of Homeruns are also displayed.

```python
#Q6: Who are the top 10 baseball players considering batting average and number of Homeruns?

#top 10 by batting average
sorted_avg = df.sort_values(by='avg', ascending=False)
top_10_avg = sorted_avg.loc[:,['name', 'avg', 'HR']][0:10]
top_10_avg
```

Then we look at the top 10 players according to number of homeruns. The data is sorted with descending order of number of homeruns and the top 10 players are filtered. Their batting average is also on display.

```python
#top 10 by number of Homeruns
sorted_hr = df.sort_values(by='HR', ascending=False)
top_10_hr = sorted_hr.loc[:,['name', 'HR', 'avg']][0:10]
top_10_hr
```

Next we will try to create a scatter plot to find if there is any correlation between the top 10 players by batting average and by the number of home runs. From the above tables we already see that these dimensions are completely disjoint. We will try to see if this can be shown with the next scatter plot. For this we will focus on the batting averages above 0.23, because the lowest batting average for top number of homeruns is 0.236. Then we mark the top 10 batting averages and the top 10 number of homeruns. To find if there is any intersection between the two, we will mark the areas in the chart corresponding with the top 10 values of both dimensions.

```python
from bokeh.models import Legend, Label

#batting averages above 0.23, because top home runs have lovest batting average of 0.236
df_avg_above_23 = df.loc[df['avg'] > 0.23]

q6 = figure(plot_width=900, plot_height=500, x_axis_label= "Batting average", y_axis_label="# Home runs",
           title="Above 0.23 batting average with number of home runs")

# add a circle renderer with a size, color, and alpha
q6.circle(df_avg_above_23['avg'], df_avg_above_23['HR'], size=5, color="navy", alpha=0.2)

# add circles for the top 10 number of homeruns
c1 = q6.circle(top_10_hr['avg'], top_10_hr['HR'], size=5, color="maroon", alpha=0.8)

#add circles for the top 10 batting average
c2 = q6.circle(top_10_avg['avg'], top_10_avg['HR'], size=5, color="green", alpha=0.8)

q6.xaxis[0].ticker.desired_num_ticks = 20

legend = Legend(items=[
    ("Top 10 # of homeruns",   [c1]),
    ("Top 10 batting averages", [c2])
], location=(0, 200))



#hightlight the best batting averages 
best_avg_box = BoxAnnotation(plot=q6, left=0.299, fill_alpha=0.15, fill_color='yellow')
#highlight best homeruns
best_hr_box = BoxAnnotation(plot=q6, bottom = 379, fill_alpha=0.15, fill_color='green')
q6.renderers.extend([best_avg_box, best_hr_box])

label = Label(x=0.298, y=382, text='Jim Rice')
q6.add_layout(label)
q6.add_layout(legend, 'left')

# show the results
show(q6)
```

From this chart we see that the top 10 batting averages and top 10 number of homeruns are disjoint. the intersection area does not contain any value. So it is difficult to determine the overall top 10 players with respecto to both the batting average and the number of homeruns. But we can approximate that one player, Jim Rice is close to the intersection, which means he has a high batting average with a high number of homeruns.
